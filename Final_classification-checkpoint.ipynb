```python
import os

train_dir = 'ODIR-5K/ODIR-5K/Training Images'
test_dir = 'ODIR-5K/ODIR-5K/Testing Images'

# Յուրաքանչյուր գրացուցակում ֆայլերի քանակը
num_train_images = len([f for f in os.listdir(train_dir) if os.path.isfile(os.path.join(train_dir, f))])
num_test_images = len([f for f in os.listdir(test_dir) if os.path.isfile(os.path.join(test_dir, f))])

print(f'Number of training images: {num_train_images}')
print(f'Number of testing images: {num_test_images}')
```

```python
def has_condn(term,text):
    if term in text:
        return 1
    else:
        return 0
```

```python
def process_dataset(data):

     # The function creates two new columns.
     #left_cataract Indicates whether the left eye has a cataract.
     #right_cataract Indicates that the right eye has a cataract.
     #It uses the helper function has_condn to check if the "cataract condition in the left/right diagnostic keywords" is present.
    data["left_cataract"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("cataract",x))
    data["right_cataract"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("cataract",x))
  
    data["LD"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("non proliferative retinopathy",x))
    data["RD"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("non proliferative retinopathy",x))

    data["LG"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("glaucoma",x))
    data["RG"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("glaucoma",x))
    
    data["LH"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("hypertensive",x))
    data["RH"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("hypertensive",x))

    data["LM"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("myopia",x))
    data["RM"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("myopia",x))
    
    data["LA"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("macular degeneration",x))
    data["RA"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("macular degeneration",x))
    
    data["LO"] = data["Left-Diagnostic Keywords"].apply(lambda x: has_condn("drusen",x))
    data["RO"] = data["Right-Diagnostic Keywords"].apply(lambda x: has_condn("drusen",x))
    
    #Filters the rows where the patient has cataracts (using data.C == 1) and stores the image IDs of the left and right eye with cataract.. .
    left_cataract_images = data.loc[(data.C ==1) & (data.left_cataract == 1)]["Left-Fundus"].values
    right_cataract_images = data.loc[(data.C == 1) & (data.right_cataract == 1)]["Right-Fundus"].values
  
    left_normal = data.loc[(data.C == 0) & (data["Left-Diagnostic Keywords"] == "normal fundus")]['Left-Fundus'].sample(350,random_state=42).values
    right_normal = data.loc[(data.C == 0) & (data["Right-Diagnostic Keywords"] == "normal fundus")]['Right-Fundus'].sample(350,random_state=42).values
    

    left_diab = data.loc[(data.C == 0) & (data.LD == 1)]["Left-Fundus"].values
    right_diab = data.loc[(data.C == 0) & (data.RD == 1)]["Right-Fundus"].values 

   
    left_glaucoma = data.loc[(data.C == 0) & (data.LG == 1)]["Left-Fundus"].values
    right_glaucoma = data.loc[(data.C == 0) & (data.RG == 1)]["Right-Fundus"].values 

    
    left_hyper = data.loc[(data.C == 0) & (data.LH == 1)]["Left-Fundus"].values
    right_hyper = data.loc[(data.C == 0) & (data.RH == 1)]["Right-Fundus"].values 
    
   
    left_myopia = data.loc[(data.C == 0) & (data.LM == 1)]["Left-Fundus"].values
    right_myopia = data.loc[(data.C == 0) & (data.RM == 1)]["Right-Fundus"].values 
    
  
    left_age = data.loc[(data.C == 0) & (data.LA == 1)]["Left-Fundus"].values
    right_age = data.loc[(data.C == 0) & (data.RA == 1)]["Right-Fundus"].values 
    
   
    left_other = data.loc[(data.C == 0) & (data.LO == 1)]["Left-Fundus"].values
    right_other = data.loc[(data.C == 0) & (data.RO == 1)]["Right-Fundus"].values 

    #Samples 350 normal images (without cataract) from both left and right eyes and stores them in arrays.
     #Selection is based on "normal found" matching diagnostic keywords.
    normalones = np.concatenate((left_normal,right_normal),axis = 0);
    cataractones = np.concatenate((left_cataract_images,right_cataract_images),axis = 0);
    diabones = np.concatenate((left_diab,right_diab),axis = 0);
    glaucoma = np.concatenate((left_glaucoma,right_glaucoma),axis = 0);
    hyper = np.concatenate((left_hyper,right_hyper),axis = 0);
    myopia = np.concatenate((left_myopia,right_myopia),axis = 0);
    age = np.concatenate((left_age,right_age),axis=0);
    other = np.concatenate((left_other,right_other),axis = 0);
    
    return normalones,cataractones,diabones,glaucoma,hyper,myopia,age,other;
```

```python
normal , cataract , diab, glaucoma , hyper , myopia , age, other = process_dataset(data);

print("Dataset stats::")
print("Normal ::" , len(normal))
print("Cataract ::" , len(cataract))
print("Diabetes ::" , len(diab))
print("Glaucoma ::" , len(glaucoma))
print("Hypertension ::" , len(hyper))
print("Myopia ::" , len(myopia))
print("Age Issues ::" , len(age))
print("Other ::" , len(other))
```

```python
from tensorflow.keras.preprocessing.image import load_img,img_to_array
dataset_dir = "preprocessed_images"
image_size=224
labels = []
dataset = []

#Function tqdm is used here to display progress bar during image processing.
def dataset_generator(imagecategory , label):#This function processes the list of images ( imagecategory ) and assigns ( label ) to them.
    for img in tqdm(imagecategory): #List of image filenames (eg normal, cataract, etc.) to process.
        imgpath = os.path.join(dataset_dir,img); #label The label assigned to this set of images (eg 0normal, 1for cataract, etc.).
        
        try:
            image = cv2.imread(imgpath,cv2.IMREAD_COLOR) #The function tries to read an image from a file using OpenCV's cv2.imread function.
                                                         #It reads the image in color mode ( cv2.IMREAD_COLOR ).
            image = cv2.resize(image,(image_size,image_size))
        except:
            continue;
        dataset.append([np.array(image),np.array(label)]);  
    random.shuffle(dataset);#After all the images in the list imagecategory have been processed and added, the dataset is shuffled to ensure that the images and labels are in random order;
    
    return dataset; #The function returns a dataset containing a pair of image data and corresponding labels.
```

```python
dataset = dataset_generator(normal,0)
dataset = dataset_generator(cataract,1)
dataset = dataset_generator(diab,2)
dataset = dataset_generator(glaucoma,3)
dataset = dataset_generator(hyper,4)
dataset = dataset_generator(myopia,5)
dataset = dataset_generator(age,6)
dataset = dataset_generator(other,7)

len(dataset)
```

```python
train_x = np.array([i[0] for i in dataset]).reshape(-1,image_size,image_size,3); #Սա հանում է բոլոր պատկերները datasetցուցակի comprehation-ը։
                                           #Images are stored as arrays and resized to fit .
#-1This is a placeholder that allows NumPy to automatically figure out the correct size based on the number of images.
#image_size, image_size This represents the height and width of the images (in this case 224x224 pixels).
#This represents the number of color channels (RGB, so 3 channels).
train_y = np.array([i[1] for i in dataset]) #This removes the tags from the dataset using datasetlist comprehation.

from sklearn.model_selection import train_test_split #
from tensorflow.keras.utils import to_categorical
x_train , x_test , y_train, y_test = train_test_split(train_x,train_y, test_size=0.2)
#This function splits the sklearn.model_selection dataset into training and testing groups.
#test_size=0.2 means that 20% of the data will be allocated to the test set, while 80% will be used for training.
y_train_cat = to_categorical(y_train,num_classes=8) #This function tensorflow.keras.utils converts an integer number of tags into single hot encoded vectors.
#For example, if the label is 2(represents "diabetes"), it will be converted to a vector like [0, 0, 1, 0, 0, 0, 0, 0].

y_test_cat = to_categorical(y_test,num_classes = 8)#Since there are 8 categories (normal, cataract, etc.), one hot coded vector will have 8 elements.
#Saves image data.
#Each input train_y is an integer between 0 and 7 representing the class the image belongs to.
#These numbers correspond to certain categories.
# 0 : Normal
#1 : Cataract
#2 : Diabetes (diabetic retinopathy)
#3 : Glaucoma
# 4. Hypertension
#5: Myopia
# 6. Age issues
#7: Other
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dropout, BatchNormalization, Dense, Flatten
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications import EfficientNetB0 
from tensorflow.keras.callbacks import LearningRateScheduler
import matplotlib.pyplot as plt

# Download the EfficientNetB0 base model without the top (classification) layer
efficientnet = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

efficientnet.trainable = False

# Define model architecture with regularization
model = Sequential()
model.add(efficientnet)  # Add EfficientNetB0 as base model

#Adding extra layers with L2 regularization and outputmodel.add(Flatten())
model.add(Dense(256, activation="relu", kernel_regularizer=l2(0.0001))) 
model.add(BatchNormalization())
model.add(Dropout(0.5))  # Dropout for regularization
model.add(Dense(256, activation="relu", kernel_regularizer=l2(0.0001)))  
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(8, activation="softmax", dtype='float32'))  

# Build the model with Adam's optimizer and cross-entropy lossmodel.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # Multi-class classification
    metrics=['accuracy']
)

# Model training
history = model.fit(
    x_train,  
    y_train_cat, 
    epochs=20,  
    batch_size=32,
    validation_data=(x_test, y_test_cat),  
    verbose=1
)

# Plotting the training and validation metrics

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, BatchNormalization, Dense, Flatten
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import LearningRateScheduler
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
import numpy as np

# Define the function to create the model
def create_model():
    efficientnet = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    efficientnet.trainable = False  # Freeze base model layers

    model = Sequential()
    model.add(efficientnet)
    model.add(Flatten())
    model.add(Dense(256, activation="relu", kernel_regularizer=l2(0.0001)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    model.add(Dense(256, activation="relu", kernel_regularizer=l2(0.0001)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    model.add(Dense(8, activation="softmax", dtype='float32'))

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# KFold cross-validation
k = 5  # Number of folds
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Placeholder to store cross-validation scores
acc_per_fold = []
loss_per_fold = []

# Convert data to numpy arrays for splitting
x_data = np.array(x_train)  # Ensure x_train is in array format
y_data = np.array(y_train_cat)

# K-fold Cross Validation
fold_no = 1
for train_index, val_index in kf.split(x_data):
    print(f'Training for Fold {fold_no} ...')
    
    # Create a new model instance
    model = create_model()
    
    # Split data for this fold
    x_train_fold, x_val_fold = x_data[train_index], x_data[val_index]
    y_train_fold, y_val_fold = y_data[train_index], y_data[val_index]
    
    # Train the model on this fold
    history = model.fit(
        x_train_fold,
        y_train_fold,
        epochs=20,
        batch_size=32,
        validation_data=(x_val_fold, y_val_fold),
        verbose=1
    )

    # Evaluate model on validation data
    scores = model.evaluate(x_val_fold, y_val_fold, verbose=0)
    print(f'Score for fold {fold_no}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1] * 100}%')
    acc_per_fold.append(scores[1] * 100)
    loss_per_fold.append(scores[0])

    # Plot metrics for this fold
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'Model accuracy for Fold {fold_no}')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'Model loss for Fold {fold_no}')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    fold_no += 1

# Display cross-validation results
print('Average accuracy per fold:', np.mean(acc_per_fold))
print('Average loss per fold:', np.mean(loss_per_fold))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np

# Assuming your model is already trained and you have x_test, y_test_cat

# Step 1: Get predicted probabilities for all classes
y_pred_proba = model.predict(x_test)

# Step 2: Binarize the true labels (convert y_test_cat into binary format)
# Example: If you have 3 classes, this will convert the labels into binary format
num_classes = y_test_cat.shape[1]
y_test_bin = label_binarize(np.argmax(y_test_cat, axis=1), classes=range(num_classes))

# Step 3: Compute the ROC curve, AUC score, and optimal threshold for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
optimal_thresholds = dict()

for i in range(num_classes):
    fpr[i], tpr[i], thresholds = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
    roc_auc[i] = roc_auc_score(y_test_bin[:, i], y_pred_proba[:, i])
    
    # Calculate Youden's J statistic and find the optimal threshold
    j_scores = tpr[i] - fpr[i]
    optimal_idx = np.argmax(j_scores)
    optimal_thresholds[i] = thresholds[optimal_idx]

# Step 4: Plot ROC curve for each class
plt.figure(figsize=(10, 7))
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f}, Optimal Threshold = {optimal_thresholds[i]:.2f})')

# Random guessing line
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()

# Step 5: Compute and print the macro-average AUC
macro_roc_auc = roc_auc_score(y_test_bin, y_pred_proba, average="macro")
print(f"Macro-average ROC AUC: {macro_roc_auc:.2f}")

# Print optimal thresholds
for i in range(num_classes):
    print(f"Optimal threshold for class {i}: {optimal_thresholds[i]:.2f}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, BatchNormalization, Dense, Flatten
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications import EfficientNetB0
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Define the optimal thresholds for each class
optimal_thresholds = [0.04, 0.02, 0.09, 0.50, 0.21, 0.74, 0.07, 0.14]

# Load the EfficientNetB0 base model without the top (classification) layer
efficientnet = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
efficientnet.trainable = False  # Freeze base model layers

# Define the model architecture with regularization
model = Sequential([
    efficientnet,
    Flatten(),
    Dense(256, activation="relu", kernel_regularizer=l2(0.0001)),
    BatchNormalization(),
    Dropout(0.5),
    Dense(256, activation="relu", kernel_regularizer=l2(0.0001)),
    BatchNormalization(),
    Dropout(0.5),
    Dense(8, activation="softmax", dtype='float32')  # 8 classes for multi-class classification
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Training the model
history = model.fit(
    x_train,
    y_train_cat,
    epochs=20,
    batch_size=32,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Plot training & validation loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot training & validation accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Predict probabilities for the test set
y_pred_probs = model.predict(x_test)

# Apply optimal thresholds to determine predicted classes
y_pred_classes = []
for probs in y_pred_probs:
    above_threshold = [i for i, prob in enumerate(probs) if prob >= optimal_thresholds[i]]
    if above_threshold:
        # Choose the class with the highest probability among those above threshold
        predicted_class = above_threshold[np.argmax([probs[i] for i in above_threshold])]
    else:
        # Default to the class with the highest probability if none meet threshold
        predicted_class = np.argmax(probs)
    y_pred_classes.append(predicted_class)

# Convert list to numpy array for compatibility with sklearn metrics
y_pred_classes = np.array(y_pred_classes)

# Generate and display confusion matrix
cm = confusion_matrix(np.argmax(y_test_cat, axis=1), y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize

# Binarize the true labels for multi-class ROC AUC calculation
y_test_bin = label_binarize(np.argmax(y_test_cat, axis=1), classes=[0, 1, 2, 3, 4, 5, 6, 7])

# Initialize plot
plt.figure(figsize=(10, 8))

# Iterate over each class and calculate ROC AUC
for i in range(8):
    # Compute the false positive rate and true positive rate
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_probs[:, i])
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')

# Plot diagonal line for random guessing
plt.plot([0, 1], [0, 1], 'k--', label='Random')

# Add plot details
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Each Class')
plt.legend(loc='lower right')
plt.show()
```
